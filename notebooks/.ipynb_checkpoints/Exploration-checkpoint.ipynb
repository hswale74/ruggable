```python
import psycopg2
import pandas as pd

# Database connection parameters
host = "localhost"
port = 5432
database = "postgres"
user = "postgres"
password = "postgres"

# Establishing the connection
conn = psycopg2.connect(
    host=host,
    port=port,
    database=database,
    user=user,
    password=password
)

# Creating a cursor object to interact with the database
cur = conn.cursor()

# Example query
query = "\dt"  # Replace with your table name

# Executing the query
cur.execute(query)

# Fetching the results
results = cur.fetchall()

# Printing the results
for row in results:
    print(row)

# Closing the cursor and connection
cur.close()
conn.close()
```

```python
def query(
    query_text, 
    host = "localhost",
    port = 5432,
    database = "postgres",
    user = "postgres",
    password = "postgres"):
    
    with psycopg2.connect(
                        host=host,
                        port=port,
                        database=database,
                        user=user,
                        password=password
                    ) as conn:
        cur = conn.cursor()
        cur.execute(query_text)
        results = cur.fetchall()
        column_names = [desc[0] for desc in cur.description]

    return_df = pd.DataFrame.from_records(results)
    return_df.columns = column_names
    return return_df
```

```python
TABLES = [
    'rugs_usa_category_map',
    'rugs_usa_color_map',
    'rugs_usa_links',
    'rugs_usa_pads_upsell',
    'rugs_usa_parent',
    'rugs_usa_variant'
]
```

```python
for table in TABLES:
    schema_query = f"""
    SELECT 
        column_name, 
        data_type, 
        is_nullable, 
        character_maximum_length, 
        numeric_precision, 
        numeric_scale
    FROM information_schema.columns
    WHERE table_schema = 'public'
    AND table_name = '{table}';
    """
    print(f"*****{table}*****")
    display(
        query(
            schema_query
        )
    )
```

```python
class PkScout:
    
    def __init__(self, table_name):
        self.table_name = table_name
        
        info_schema_query = f"""
            SELECT 
            column_name, 
            data_type, 
            is_nullable, 
            character_maximum_length, 
            numeric_precision, 
            numeric_scale
            FROM information_schema.columns
            WHERE table_schema = 'public'
            AND table_name = '{self.table_name}';
        """

        self.info_schema_df = query(info_schema_query)
        self.table_row_count = int(query(
            f"Select count(1) as ct from {self.table_name}"
        )['ct'].iloc[0])

    def get_column_cardinalities(self):
        column_cardinalities = []
        for column in self.info_schema_df['column_name']:
            column_cardinality = int(
                query(
                    f'Select count(distinct "{column}") as ct from {self.table_name}'
                )['ct'].iloc[0]
            )
            
            column_cardinalities.append((column, column_cardinality))

        return sorted(column_cardinalities, key=lambda x: x[1], reverse=True)

    def find_pk(self):
        column_cardinalities = self.get_column_cardinalities()

        joint_cardinalities = []
        for index, (column, cardinality) in enumerate(column_cardinalities):
            if cardinality == self.table_row_count:
                print(f"{column} is a candidate: Same cardinality as table row count")
            else:
                for i in range(index, len(column_cardinalities)-1):
                    joint_columns = [f'"{column}"'] + [
                        f'"{name}"' for name, _ in column_cardinalities[index+1:i+2]
                    ]
                    joint_cardinality = query(
                        f"""
                            Select count(1) as ct
                            from (
                                Select distinct {','.join(joint_columns)}
                                from {self.table_name}
                            ) as foo
                        """
                    )['ct'].iloc[0]

                    if joint_cardinality == self.table_row_count:
                        print(f"{joint_columns} are a candidate: Same cardinality as table row count")

                    print(
                        (
                                tuple([x.replace('"','') for x in joint_columns]),
                                joint_cardinality
                    
                        )
                    )
                    joint_cardinalities.append(
                        (
                            tuple([x.replace('"','') for x in joint_columns]),
                            joint_cardinality
                        )
                    )
                    

            
```
